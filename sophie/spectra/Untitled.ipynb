```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('notebook')
```

```python
wav, flux = np.loadtxt("SOPHIE_spliced_vcor.obs", unpack=True)

adel_wav, adel_flux = np.loadtxt('3850-7000_Adelman2011.prf', unpack=True, skiprows=6763)
new_wav, new_flux = np.loadtxt('3850-7000_New.prf', unpack=True, skiprows=13948)
```

```python
plt.plot(wav, flux)
plt.xlim(3940, 3990)
```

```python

plt.figure(figsize=[20,7])

# plt.plot(wav, flux, c='black', lw=0.8, label='Data')
plt.plot(adel_wav, adel_flux, lw=1., c='blue', label='Old abundances')
plt.plot(new_wav, new_flux, lw=0.7, c='red', label='Richard')
plt.xlim(3900, 4000)
# plt.ylim(0.8, 1.01)
plt.legend()
```
