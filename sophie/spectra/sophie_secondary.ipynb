```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('science')
```

```python
wav, flux = np.loadtxt("SOPHIE_spliced_vcor.obs", unpack=True)

adel_wav, adel_flux = np.loadtxt('3850-7000_Adelman2011.prf', unpack=True, skiprows=6763)
new_wav, new_flux = np.loadtxt('3850-7000_New.prf', unpack=True, skiprows=13948)
```

```python
overleaf_path = '/Users/daniel/Dropbox (Sydney Uni)/Apps/Overleaf/adra/figs/'
```

```python
fig, axes = plt.subplots(1, 1, figsize=[10/3, 2.5])

plt.plot(wav, flux, lw=0.8, c='black', label='SOPHIE')
plt.plot(adel_wav, adel_flux, c='C0', label='Synthetic')

plt.axvline(4481.1, ymin=0.82, ymax=0.87, c='C3')
plt.text(4481.1, 1.07, 'Mg II', fontsize=6, rotation=90, horizontalalignment='center',
         verticalalignment='center')
plt.xlim(4465, 4495)
# plt.xlim(3920, 3950)
plt.ylim(0.5, 1.1)

plt.legend(fontsize=7)

plt.xlabel(f"Wavelength [$\AA$]")
plt.ylabel('Normalised flux')

plt.savefig(overleaf_path+'Mg4481_id_dan.pdf', dpi=300, bbox_inches='tight')
```
