```python
import matplotlib.pyplot as plt
import numpy as np
import pymc3 as pm
import exoplanet as xo
import starry

np.random.seed(12)
starry.config.quiet = True
```

```python
from starry import Map
```

```python
A = starry.Primary(starry.Map(udeg=2), 
                   r=2, 
                   m=1.0)
A.map[1:] = [0.5, 0.25]
```

```python
b = starry.Secondary(
    starry.Map(ydeg=10, inc=86.3789, obl=30.0, L=0.1),
    r=4,
    m=1,
    a=87.567,
#     porb=1.,
    prot=1.0,
    ecc=0.427,
    t0=0.0,
    inc=86.3789,
    Omega=22.51266,
)
```

```python
sys = starry.System(A, b)
```

```python
npts = 200
t = np.linspace(-0.5, 0.5, npts)
x, y, z = sys.position(t)
```

```python
flux = sys.flux(t)
plt.plot(t, flux.eval())
```

```python
sys.show?
```

```python
res = 200
theta_pri = 360.0 / sys.primary.prot * (t - sys.primary.t0) - sys.primary.theta0
theta_sec = [360.0 / sec.prot * (t - sec.t0) - sec.theta0 for sec in sys.secondaries]
img = np.array(
    [sys.primary.map.render(theta=theta_pri, res=res)]
    + [
        sec.map.render(theta=theta_sec[i], res=res)
        for i, sec in enumerate(sys.secondaries)
    ]
)
```

```python
from matplotlib.animation import FuncAnimation
from IPython.core.display import HTML

# Set up the plot
fig, ax = plt.subplots(1, figsize=(12,12))
ax_xz = fig.add_axes([0.275, 0.8, 0.2, 0.2])
ax_xz.annotate(
    "Top", fontsize=12, xy=(0, 0), xycoords="axes fraction", ha="left", va="bottom"
)
ax_zy = fig.add_axes([0.525, 0.8, 0.2, 0.2])
ax_zy.annotate(
    "Side", fontsize=12, xy=(0, 0), xycoords="axes fraction", ha="left", va="bottom"
)
ax_lc = fig.add_axes([0.125, 0.05, 0.775, 0.2])

xz = [None] + [None for sec in sys.secondaries]
xy = [None] + [None for sec in sys.secondaries]
zy = [None] + [None for sec in sys.secondaries]
circ = [None] + [None for sec in sys.secondaries]
maps = [sys.primary.map] + [sec.map for sec in sys.secondaries]
radii = np.array([sys.primary.r] + [sec.r for sec in sys.secondaries])

for axis, arrs in zip([ax, ax_xz, ax_zy], [(x.eval(), y.eval()), (x.eval(), z.eval()), (z.eval(), y.eval())]):
    axis.axis("off")
    R = 1.2 * max(-np.min(arrs), np.max(arrs))
    axis.set_xlim(-R, R)
    axis.set_ylim(-R, R)

# Plot the light curve
ax_lc.plot(t, flux.eval(), "k-")
(lc,) = ax_lc.plot(t[0], flux.eval()[0], "o", color="k")
ax_lc.axis("off")

# Plot the first frame
for i, xi, yi, zi, map, r in zip(range(1 + len(sys.secondaries)), x.eval(), y.eval(), z.eval(), maps, 
                                 [radii[0].eval(), radii[1].eval()]):

    # Orbit outlines
    ax_xz.plot(xi, zi)
    ax_zy.plot(zi, yi)

    # Body positions
    xz[i] = ax_xz.scatter(xi[0], zi[0])
    zy[i] = ax_zy.scatter(zi[0], yi[0])

    # Maps
    extent = np.array([xi[0], xi[0], yi[0], yi[0]]) + np.array([-1, 1, -1, 1]) * r
    xy[i] = ax.imshow(
        img[i].eval()[0],
        origin="lower",
        cmap="plasma",
        extent=extent,
        clip_on=False,
        zorder=zi[0],
    )
    circ[i] = plt.Circle(
        (xi[0], yi[0]), r, color="k", fill=False, zorder=zi[0] + 1e-3, lw=0.
    )
    ax.add_artist(circ[i])

# Animation
def updatefig(k):
    for i, xi, yi, zi, map, r in zip(
        range(1 + len(sys.secondaries)), x.eval(), y.eval(), z.eval(), maps, [radii[0].eval(), radii[1].eval()]
    ):
        xz[i].set_offsets((xi[k], zi[k]))
        zy[i].set_offsets((zi[k], yi[k]))
        xy[i].set_extent(
            np.array([xi[k], xi[k], yi[k], yi[k]]) + np.array([-1, 1, -1, 1]) * r
        )
        xy[i].set_zorder(zi[k])
        xy[i].set_data(img[i].eval()[k])
        circ[i].center = (xi[k], yi[k])
        circ[i].set_zorder(zi[k] + 1e-3)
    lc.set_xdata(t[k])
    lc.set_ydata(flux.eval()[k])
    return xz + xy + zy + circ + [lc]
```

```python
ani = FuncAnimation(fig, updatefig, interval=30, blit=False, frames=len(t))
plt.close()
display(HTML(ani.to_html5_video()))
```

```python

with pm.Model() as model:
    A = starry.Primary(starry.Map(udeg=2, L=1.0), r=1.0, m=1.0)
    A.map[1] = 0.40
    A.map[2] = 0.25

A.map.show(theta=np.linspace(0, 360, 50))
```

```python
with model:
    B = starry.Secondary(
        starry.Map(ydeg=5, inc=75.0, L=0.1),
        r=0.7,
        m=0.7,
        porb=1.00,
        prot=0.625,
        t0=0.15,
        inc=75.0,
    )
    B.map.add_spot(amp=-0.075, lat=-30, lon=60)

B_true_map = B.map[1:, :].eval()
B.map.show(theta=np.linspace(0, 360, 50))
```

```python
with model:
    sys = starry.System(A, B)

# Compute the light curve
t = np.linspace(-2.5, 2.5, 1000)
flux_true = sys.flux(t).eval()

# Add noise
sigma = 0.00025
flux = flux_true + sigma * np.random.randn(len(t))
```

```python
fig, ax = plt.subplots(1, figsize=(12, 5))
ax.plot(t, flux, "k.", alpha=0.3, ms=2)
ax.plot(t, flux_true)
ax.set_xlabel("time [days]", fontsize=24)
ax.set_ylabel("relative flux", fontsize=24);
```

```python
with model:

    # The Ylm coefficients with a simple gaussian prior
    ncoeff = B.map.Ny - 1
    mu = np.zeros(ncoeff)
    cov = 1e-2 * np.eye(ncoeff)
    Ylm = pm.MvNormal("Ylm", mu, cov, shape=(ncoeff,))
    B.map[1:, :] = Ylm

    # Compute the flux
    flux_model = sys.flux(t=t)

    # Track some values for plotting later
    pm.Deterministic("flux_model", flux_model)

    # Save our initial guess
    flux_model_guess = xo.eval_in_model(flux_model)

    # The likelihood function assuming known Gaussian uncertainty
    pm.Normal("obs", mu=flux_model, sd=sigma, observed=flux)
```
